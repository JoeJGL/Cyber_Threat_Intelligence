## pypdf

```python
from pypdf import PdfReader
```

```python
reader = PdfReader(path)
```

```python
text_pypdf = ""
for p in reader.pages :
    text_pypdf += p.extract_text()
print(text_pypdf)
```

## Cleaning

```python
import re

def clean_pdf_text(text):

    # Retirer les headers/footer connus
    text = text.replace("TLP:CLEAR", " ")

    # Retirer pagination
    text = re.sub(r"\b\d+/\d+\b", " ", text)

    # Retirer références inline
    text = re.sub(r"\[\s*\d+(?:,\s*\d+)*\s*\]", " ", text)

    # Retirer URLs
    text = re.sub(
        r"https?://\S+",
        " ",
        text
    )

    # Retirer césures PDF
    text = re.sub(r"-\n", "", text)

    # Normalisation espaces
    text = re.sub(r"\s+", " ", text)

    return text.strip()
```

```python
cleaned_text_pypdf = clean_pdf_text(text_pypdf)
print(cleaned_text_pypdf)
```

```python
try:
    with open("cleaned_text_pypdf.txt", "x", encoding="utf-8") as f:
        f.write(cleaned_text_pypdf)
except FileExistsError:
    print("file.txt already exists, exclusive creation aborted.")
```
